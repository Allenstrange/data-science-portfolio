# Gold Price Analysis Project

This project analyzes the historical gold prices in various currencies to uncover trends, impacts of major global events, and seasonal variations.

**The datasets contain:**
- `annual_gold_rate.csv`: Annual average gold rates in different currencies.
- `daily_gold_rate.csv`: Daily gold rates over the years.

**Questions to Answer:**
1. Plot a graph visualizing the gold price in different currencies.
2. Plot a graph representing gold rates around the years of 2008's global economic crisis.
3. Effect of COVID-19 on gold rates (Visualization).
4. In which months are the gold rates lowest in that year (Examining this trend for 2014-2021).


## 1. Import Libraries and Load Data
Let's start by importing the necessary libraries and loading the datasets into pandas DataFrames.

```python
# Importing necessary libraries for data manipulation and visualization
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import datetime as dt

# Loading the datasets
annual_gold = pd.read_csv('annual_gold_rate.csv')
daily_gold = pd.read_csv('daily_gold_rate.csv')

# Display the first few rows of the annual dataset
annual_gold.head()
```

## 2. Data Cleaning
We need to handle missing values in some of the currency columns by filling them with the mean value of the respective column.

```python
# Fill missing values with the mean of the column (cast to integer)
annual_gold['INR'].fillna(int(annual_gold['INR'].mean()), inplace=True)
annual_gold['AED'].fillna(int(annual_gold['AED'].mean()), inplace=True)
annual_gold['CNY'].fillna(int(annual_gold['CNY'].mean()), inplace=True)

daily_gold['CNY'].fillna(int(daily_gold['CNY'].mean()), inplace=True)
```

## Question 1: Visualizing the Gold Price in Different Currencies
To understand the overall trend of gold prices over the decades, we will plot line charts for each currency available in the dataset.

```python
# List of currencies to visualize
currencies = ['USD', 'EUR', 'GBP', 'INR', 'CNY', 'AED']

# Plotting the historical trend for each currency
for currency in currencies:
    fig = px.line(
        annual_gold, 
        x='Date', 
        y=currency, 
        labels={currency: 'Gold Prices'},
        title=f'Gold Prices In {currency} Through The Years'
    )
    fig.show()
```

## Question 2: Gold Rates Around the 2008 Global Economic Crisis
The 2008 financial crisis had a profound impact on global markets. We will look at the gold rates spanning from roughly 2005 to 2011 to see how prices reacted before, during, and after the crisis.

```python
# Filter the dataset for the years surrounding the 2008 crisis (2005-2011)
# 2005 is index 27, 2011 is index 33 in the annual_gold dataframe.
crisis_years = annual_gold[(annual_gold['Date'] >= 2005) & (annual_gold['Date'] <= 2011)]

# Visualizing the impact of the crisis for each currency
for currency in currencies:
    fig = px.bar(
        crisis_years, 
        x='Date', 
        y=currency, 
        labels={currency: 'Gold Prices'},
        title=f'Gold Prices In {currency} Around the 2008 Economic Crisis'
    )
    fig.update_layout(xaxis_type='category') # Display years discretely
    fig.show()
```

## Question 3: Effect of COVID-19 on Gold Rates
The COVID-19 pandemic started significantly affecting global markets in early 2020. We will analyze the trend of gold prices leading up to and during the pandemic (2017 to 2021) to visualize its impact as investors flocked to "safe-haven" assets.

```python
# Filter the dataset for the years relevant to the COVID-19 pandemic (2017-2021)
covid_years = annual_gold[(annual_gold['Date'] >= 2017) & (annual_gold['Date'] <= 2021)]

# Visualizing the impact of COVID-19 for each currency
for currency in currencies:
    fig = px.bar(
        covid_years, 
        x='Date', 
        y=currency, 
        labels={currency: 'Gold Prices'},
        title=f'Effect of COVID-19 on Gold Prices In {currency} (2017-2021)'
    )
    fig.update_layout(xaxis_type='category') # Display years discretely
    fig.show()
```

## Question 4: Analyzing Monthly Minimums (2014-2021)
For this analysis, we want to find out which month typically records the lowest gold rate in a given year. We will examine the daily gold data from 2014 to 2021.

```python
# Convert the 'Date' column to datetime format for easier extraction of year and month
daily_gold['Date'] = pd.to_datetime(daily_gold['Date'])

# Extract 'year' and 'month' into separate columns
daily_gold['year'] = daily_gold['Date'].dt.year
daily_gold['month'] = daily_gold['Date'].dt.month

# Map numerical month values to their string names for better visualization
month_map = {
    1: 'January', 2: 'February', 3: 'March', 4: 'April', 
    5: 'May', 6: 'June', 7: 'July', 8: 'August', 
    9: 'September', 10: 'October', 11: 'November', 12: 'December'
}
daily_gold['month_name'] = daily_gold['month'].replace(month_map)

# Set index to 'year' to facilitate yearly queries
daily_gold.set_index(['year'], inplace=True)
daily_gold.head()
```

```python
def plot_minimum_months(currency):
    """
    Finds the month with the lowest gold rate for each year from 2014 to 2021
    and plots a bar chart representing these lowest months.
    """
    years_of_interest = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]
    lowest_months = []
    
    for year in years_of_interest:
        # Retrieve data for the specific year
        yearly_data = daily_gold.loc[year]
        
        # Find the minimum value for the currency in that year
        min_rate = yearly_data[currency].min()
        
        # Get the corresponding month name for the minimum value
        # Using .iloc[0] to handle cases with multiple days having the exact same minimum rate
        month_with_min = yearly_data[yearly_data[currency] == min_rate]['month_name'].iloc[0]
        lowest_months.append(month_with_min)
        
    # Create the bar chart
    fig = go.Figure([go.Bar(x=years_of_interest, y=lowest_months)])
    
    # Configure layout
    fig.update_xaxes(title_text="Years", type='category')
    fig.update_yaxes(
        title_text="Months", 
        categoryorder='array', 
        categoryarray=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
    )
    fig.update_layout(title_text=f'Month with the Lowest Gold Rate in {currency} (2014-2021)')
    fig.show()

# Run the function for all defined currencies
for currency in currencies:
    print(f'Analyzing {currency}...')
    plot_minimum_months(currency)
```
